```python
#Bibliotecas usadas

import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Cod de visualisação de Dados

Base_Dados = pd.read_excel('Dados_Empresas_Energia.xlsx')

Base_Dados.head()
```

```python
Base_Dados.set_index( 'Data', inplace=True )

Base_Dados.head()
```

```python
# Grafico

# Estilo do gráfico
plt.style.use('seaborn-darkgrid')

# Tamanho
plt.figure( figsize=(16, 7))

# Titulo
plt.title('Análise ações de empresas de Energia', loc='left', fontsize=18, fontweight=0)

# Plot/visualização da Petrobras
# Tudo edição/parametros do grafico (color='#008c4a', linewidth=4, alpha=0.7)

plt.plot( Base_Dados.index, Base_Dados['Petrobras'], color='#008c4a', linewidth=4, alpha=0.7)
#Texto Petrobras para incluir no grafico
plt.text( Base_Dados.index[-1], Base_Dados['Petrobras'].tail(1), 'Petrobras', color='#008c4a', size='large', horizontalalignment='left')

# Plot das Outras Colunas
for Coluna in Base_Dados.columns[1:]:
    # Plot das outras colunas
    plt.plot(Base_Dados.index, Base_Dados[Coluna], color='gray', linewidth=2, alpha=0.7)
    # Texto para incluir no gráfico
    plt.text(Base_Dados.index[-1], Base_Dados[Coluna].iloc[-1], Coluna, color='gray', size='large', horizontalalignment='left')

# Labels
plt.xlabel('Período')
plt.ylabel('Preço de Fechamento (R$)');
```
